# Fetching daily data


Fetching the spectra from dailies by tileid, night and fiber
Here we fetch the cumulative coadd for a given tile.

```python
fiber = 201 
night = '20230125'
tileid = 83203
SP = desir.get_specs(fiber=fiber,
                               coadd_type='tiles/cumulative',
                               night=night,
                               tileid=tileid)[0]
print (SP)
```

```python
drp.plot(SP)
```

For a subset of dailies that were fitted it is possible to also overplot the RVSpec models
For this you need the same info as  to fetch the data, but you may also need to fetch run 
,i.e. the string used to identify the rvspecfit run (i.e. 220309 for 
data.desi.lbl.gov/desi/science/mws/redux/daily/rv_output/220309  )

```python
fiber = 201 
tileid = 82403
night = '20211119'
SP = desir.get_specs(fiber=fiber,
                    coadd_type='tiles/cumulative',
                  night=night,
                    tileid=tileid)

SPM = desir.get_rvspec_models(fiber=fiber,
                               coadd_type='tiles/cumulative',
                               night=night,
                               tileid=tileid,
                            run='220309')
```
