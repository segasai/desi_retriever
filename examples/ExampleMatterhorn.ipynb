# Matterhorn

Fetching spectra by survey/subsurvey, healpix and targetid 


```python
SP = get_specs(survey='sv1',
               program='dark',
                uniqpix=8516,
               targetid=39627652591521181)[0]
print(SP)
```

Fetching the variances and fibermaps

```python
SP = get_specs(survey='sv1',
               program='dark',
               uniqpix=8516,
               targetid=39627652591521181,
               fibermap=True,
               ivar=True)[0]
print(SP)
```

## Fetching the corresponding rvspec models

```python
SPM = get_rvspec_models(survey='sv1',
                        program='dark',
                        uniqpix=8516,
                        targetid=39627652591521181)[0]
```

```python
print(SPM)
```

```python
desi_plot(SP, model=SPM)
```
